```python
s1 = np.random.uniform(0,1,10**6)
s2 = np.random.uniform(0,1,10**6)
## flat spins aligned mostly with z direction
cos_theta = np.random.triangular(0.86, 1, 1 ,10**6)
theta = np.arccos(cos_theta)
phi = np.random.uniform(0, 2*np.pi , 10**6)
```

```python
s1x = s1*np.sin(theta)*np.cos(phi)
s1y = s1*np.sin(theta)*np.sin(phi)
s1z = s1*cos_theta
```

```python
sns.distplot(s1)
plt.show()
```

```python
sns.distplot((cos_theta))
plt.show()
```

```python
sns.distplot(np.degrees(theta))
plt.show()
```

```python
sns.distplot(s1z)
plt.show()
```

```python
sns.distplot(np.sqrt(s1x**2 + s1y**2))
plt.show()
```
